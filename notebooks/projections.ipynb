```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 6))
ax = plt.axes(projection=ccrs.PlateCarree())
gl = ax.gridlines(draw_labels=True,linewidth=2,color='gray',alpha=0.2,linestyle='--')
ax.coastlines()
```

```python
plt.figure(figsize=(12, 6))

ax = [
    plt.subplot(131,projection=ccrs.LambertConformal()),
    plt.subplot(132,projection=ccrs.LambertConformal(central_longitude=60)),
    plt.subplot(133,projection=ccrs.LambertConformal(central_longitude=180))
]
for x in range(0, 3, 1):
  gl = ax[x].gridlines(draw_labels=True,linewidth=2,color='gray',alpha=0.2,linestyle='--')
  ax[x].coastlines()
  ax[x].set_extent((-150,-10 , 0,90 ),ccrs.PlateCarree())
```

```python
plt.figure(figsize=(12, 6))

ax = [
    plt.subplot(131,projection=ccrs.LambertConformal()),
    plt.subplot(132,projection=ccrs.LambertConformal(central_latitude=20)),
    plt.subplot(133,projection=ccrs.LambertConformal(central_latitude=60))
]
for x in range(0, 3, 1):
  gl = ax[x].gridlines(draw_labels=False,linewidth=2,color='gray',alpha=0.2,linestyle='--')
  ax[x].coastlines()
  ax[x].set_extent((-130,-60 , 15,60 ),ccrs.PlateCarree())
  ax[x].add_feature(cfeature.STATES, edgecolor='black', linewidth=0.5)
```

```python
plt.figure(figsize=(12, 6))

ax = [
    plt.subplot(131,projection=ccrs.Stereographic(central_longitude=-100,central_latitude=20)),
    plt.subplot(132,projection=ccrs.Stereographic(central_longitude=-100,central_latitude=45)),
    plt.subplot(133,projection=ccrs.Stereographic(central_longitude=-100,central_latitude=70))
]
for x in range(0, 3, 1):
  gl = ax[x].gridlines(draw_labels=True,linewidth=2,color='gray',alpha=0.2,linestyle='--')
  ax[x].coastlines()
  ax[x].set_extent((-160,-60 , 15,90 ),ccrs.PlateCarree())
  ax[x].add_feature(cfeature.STATES, edgecolor='black', linewidth=0.5)
```
